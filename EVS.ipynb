# Distribution Analysis

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Model Year'], kde=True)
plt.title('Distribution of Model Year')
plt.xlabel('Model Year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Electric Range'], kde=True)
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(15, 4))
sns.boxplot(x='Clean Alternative Fuel Vehicle (CAFV) Eligibility', y='Electric Range', data=df)
plt.title('Electric Range by CAFV Eligibility')
plt.xlabel('CAFV Eligibility')
plt.ylabel('Electric Range')
plt.show()
```


# Trend Analysis


```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Model Year', y='Electric Range', data=df)
plt.title('Electric Range Over the Years')
plt.xlabel('Model Year')
plt.ylabel('Electric Range')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Model Year', y='Base MSRP', data=df)
plt.title('Base MSRP Over the Years')
plt.xlabel('Model Year')
plt.ylabel('Base MSRP')
plt.show()
```

# Geographical Analysis

```python
top_10_counties = df['County'].value_counts().nlargest(10).index

plt.figure(figsize=(10, 6))
sns.countplot(y='County', data=df[df['County'].isin(top_10_counties)], order=top_10_counties)
plt.title('Distribution of EVs by County')
plt.xlabel('Count')
plt.ylabel('County')
plt.show()
```

########Forecast trends for the next 5 years based on past data

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv('EVS.csv')
```

```python
df['ds'] = pd.to_datetime(df['Model Year'], format='%Y')
df_range = df[['ds', 'Electric Range']].rename(columns={'Electric Range': 'y'})
df_msrp = df[['ds', 'Base MSRP']].rename(columns={'Base MSRP': 'y'})
```

```python
df_range.dropna(inplace=True)
```

```python
df_msrp.dropna(inplace=True)


```

```python
model_range = Prophet()
```

```python
model_range.fit(df_range)

```

```python
future_range = model_range.make_future_dataframe(periods=5 * 365)
forecast_range = model_range.predict(future_range)
```

```python
plt.figure(figsize=(10, 6))
model_range.plot(forecast_range)
plt.title('Forecast of Electric Range for the Next 5 Years')
plt.xlabel('Year')
plt.ylabel('Electric Range')
plt.show()
```

```python
model_range.plot_components(forecast_range);
```

```python
model_msrp = Prophet()
```

```python
model_msrp.fit(df_msrp)
```

```python
future_msrp = model_msrp.make_future_dataframe(periods=5 * 365)
forecast_msrp = model_msrp.predict(future_msrp)
```

```python
plt.figure(figsize=(10, 6))
model_msrp.plot(forecast_msrp)
plt.title('Forecast of Base MSRP for the Next 5 Years')
plt.xlabel('Year')
plt.ylabel('Base MSRP')
plt.show()
```

```python
model_msrp.plot_components(forecast_msrp);
```
